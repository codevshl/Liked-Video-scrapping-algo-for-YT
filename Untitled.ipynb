```python
from googleapiclient.discovery import build

# Initialize the YouTube Data API client
youtube = build("youtube", "v3", credentials=credentials)

def get_liked_videos():
    videos_list = []
    next_page_token = None

    while True:
        # Get the liked videos from the "videos" endpoint
        liked_response = youtube.videos().list(
            part="snippet",
            myRating="like",
            maxResults=50,  # You can adjust this to fetch more or fewer videos per request
            pageToken=next_page_token,
        ).execute()

        # Extract video details and add them to the list
        for item in liked_response.get("items", []):
            video_id = item["id"]
            title = item["snippet"]["title"]
#             description = item["snippet"]["description"]
            channel_title = item["snippet"]["channelTitle"]
            video_url = f"https://www.youtube.com/watch?v={video_id}"
            
            video_info = {
                "title": title,
#                 "description": description,
                "channel_title": channel_title,
                "video_url": video_url
            }

            videos_list.append(video_info)

        # Check if there are more liked videos
        next_page_token = liked_response.get("nextPageToken")
        if not next_page_token:
            break

    return videos_list

# Call the function to get the liked videos with all details
liked_videos = get_liked_videos()

# Print the list of liked videos with all details
for index, video_details in enumerate(liked_videos, 1):
    print(f"{index}. Title: {video_details['title']}")
#     print(f"   Description: {video_details['description']}")
    print(f"   URL: {video_details['video_url']}\n")
    print(f"   Channel: {video_details['channel_title']}")
#     print(f"   Published At: {video_details['published_at']}")
    # Print more details if needed
    print("-" * 30)
```
